# WORLD.jl

WORLD.jl is julia wrapper for [WORLD](http://ml.cs.yamanashi.ac.jp/world/), a high-quality speech analysis, modification and synthesis system written in C++. This notebook demonstrates how WORLD.jl works.

```julia
using PyPlot
using WAV
using WORLD
```

```julia
# Reading a speech signal
filepath = joinpath(Pkg.dir("WORLD"), "test", "data", "test16k.wav")
x, fs = wavread(filepath)
x = vec(x) # monoral
fs = int(fs)

# Visualize the speech signal in time-domain
figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(1:endof(x), x, label="a speech signal")
xlim(1, endof(x))
xlabel("sample")
legend()
```

```julia
# Basic setup for WORLD-based analysis. (*This is not necessary but might be useful. The composite type `World` just holds sample frequency 
# and frame period that are used during analysis. You can call WORLD functions directly*) 
period = 5.0
w = World(fs, period)
```

```julia
# Fundamental frequency (f0) estimation by DIO
opt = DioOption(40.0, 700, 2.0, period, 4)
f0, timeaxis = dio(w, x, opt=opt)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(timeaxis, f0, label="F0 trajectory estimated by DIO", linewidth="2")
xlabel("Time [sec]")
ylabel("Frequency  [Hz]")
legend()
```

```julia
f0_by_dio = copy(f0)

# F0 refinement by StoneMask
f0 = stonemask(w, x, timeaxis, f0)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(timeaxis, f0_by_dio, label="F0 trajectory estimated by DIO", linewidth="2")
plot(timeaxis, f0, label="Refined f0 by StoneMask", linewidth="2")
xlabel("Time [sec]")
ylabel("Frequency  [Hz]")
legend()
```

```julia
# Spectral envelope estimatino by Star
spectrogram = star(w, x, timeaxis, f0)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(log(spectrogram), origin="lower", aspect="auto")
colorbar()
```

```julia
# Spectral envelope estimatino by CheapTrick
spectrogram = cheaptrick(w, x, timeaxis, f0)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(log(spectrogram), origin="lower", aspect="auto")
colorbar()
```

```julia
# Residual spectrogram estimation by Platinum
residual = platinum(w, x, timeaxis, f0, spectrogram)
    
figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(residual, origin="lower", aspect="auto")
colorbar()
```

```julia
# Sysnthesis from f0, spectral envelope and residual spectrogram.
y = synthesis(w, f0, spectrogram, residual, length(x))

println("Mean error: ", mean(abs(y-x))/maxabs(x))

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(1:endof(y), y, "r-+", label="reconstructed speech signal")
plot(1:endof(x), x, label="original speech signal")
xlim(1, endof(y))
xlabel("sample")
legend()
```

```julia
# Aperiodicity ratio estimation
aperiodicity = aperiodicityratio(w, x, f0, timeaxis)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(log(aperiodicity), origin="lower", aspect="auto")
colorbar()
```

```julia
# Sysnthesis from f0, spectral envelope and aperiodicity ratio.
ya = synthesis_from_aperiodicity(w, f0, spectrogram, aperiodicity, length(x))

println("Mean error: ", mean(abs(ya-x))/maxabs(x))

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(1:endof(ya), ya, "r-+", label="reconstructed speech signal from aperiodicity ratio")
plot(1:endof(x), x, label="original speech signal")
xlim(1, endof(ya))
xlabel("sample")
legend()
```

## Conversion between spectrum envelope and mel-cepstrum

```julia
α = 0.41
fftlen = get_fftsize_for_cheaptrick(fs)
order = 40

mc = sp2mc(spectrogram, order, α)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(mc, origin="lower", aspect="auto")
colorbar()
```

```julia
approximate_spectrogram = mc2sp(mc, α, fftlen)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
imshow(log(approximate_spectrogram), origin="lower", aspect="auto")
colorbar()
```
